```python
n = int(input("Enter number of rows: "))
for i in range(1,n+1):
    for j in range (1,2*n):
        if (i==n or i+j==n+1 or j-i==n-1):   #printing stars at last row, for left slant (col+row=n+1), for right (col-row=n-1)
            print("*",end="")
        else:
            print(end=" ")
    print()
```

```python
n = int(input("Enter number of rows: "))
k=2
for i in range(1,n+1):
    for j in range (1,2*n):
        if (i+j==n+1 or j-i==n-1):
            print("*",end="")
        elif (i==n and j!=k):
            print("*",end="")
            k=k+2
        else:
            print(end=" ")
    print()
```

```python
n = int(input("Enter number of rows: "))
for i in range(1,n+1):
    for j in range(1,n-i+1):  #left orange triangle
        print(end=" ")
    for j in range(i,0,-1):   #green triangle
        print(j,end=" ")
    for j in range(2,i+1):    #right orange triangle
        print(j,end=" ")
    print()
```
